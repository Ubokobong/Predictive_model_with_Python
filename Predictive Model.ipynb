```python
# start by importing neccesary data science libraries in python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#load and read data
df = pd.read_csv('Kerala.csv')
df.head(5)
```

```python
#Exploring the data
df.info()
```

```python
#changing the output column from yes/no to 0/1 helps to train the model better
```

```python
df['FLOODS'].replace(['YES','NO'],[1,0], inplace=True)
df.head(5)
```

```python
#next we do a feature selection to pick features that contribute the most to the output
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
#group the independent variables as X and output as Y
```

```python
X= df.iloc[:,1:14]   
Y= df.iloc[:,-1]  
```

```python
best_features= SelectKBest(score_func=chi2, k=3) #3 to select top 3
fit= best_features.fit(X,Y)
```

```python
#create a dataframe and show the features and scores
```

```python
df_scores= pd.DataFrame(fit.scores_)
df_columns= pd.DataFrame(X.columns)
```

```python
features_scores= pd.concat([df_columns, df_scores], axis=1)
features_scores.columns= ['Features', 'Score']
features_scores.sort_values(by = 'Score')
```

```python
#the top 3 features that relates better to the output is SEP,JUN,JUL
```

```python
#Building Model - group the independent variables into X and dependent into Y
```

```python
#Split the dataset into X and Y:

X= df[['SEP', 'JUN', 'JUL']]  
Y= df[['FLOODS']]  

#Split the dataset into train and test:

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.4,random_state=100)

#Create a logistic regression body:

logreg= LogisticRegression()
logreg.fit(X_train,y_train)

#the likelihood of a flood is predicted using the logistic regression body we created:

y_pred=logreg.predict(X_test)
print (X_test) #test dataset
print (y_pred) #predicted values
```
